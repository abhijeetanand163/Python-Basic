**Tensorflow variables : **
A variable is a tensor whose value is initiated and then typically change as program runs

```
import tensorflow as tf
```

```
# Forward pass takes two input w and x and do matrix multlipication on it !
def forward_pass(w,x):
  return tf.matmul(w,x)
```

In the code below **w** is the variable. It's name is  "weights''.

It's not initialized here because tensor flow is lazy evaluation framework. Here we are only building the graph. When it is initialized, it will be initialized with truncated normal initializer. It's initialize the tensor with random number which are distributed according to gaussian normal distribution with 'zero' mean and 'unit' variance. 

**w** is trainable , it is it can be changed during training. 

```
def train_loop(x, niter = 5):
  with tf.variable_scope("model", reuse = tf.AUTO_REUSE): # here we are telling tf to reuse variable each time instead of creating variable each time !
    w = tf.get_variable("weights",                     # initialize w 
                   shape = (1,2), # 1 X 2 matrix
                   initializer = tf.truncated_normal_initializer(), # in neural network initializing with random normal initializer is a normal case
                   trainable = True)
  preds = []  # Training loop of 5 updates to weights.
  for k in xrange(niter):
    preds.append(forward_pass(w,x))
    w = w + 0.1 # gradient update !
  return preds
```

**w** is **1 X 2** matrix and x is **2 X 3** matrix (defined below), multiplying both of them gives **1 X 3** matrix !!

```
with tf.Session() as sess:
  preds = train_loop([[3.2 , 5.1 ,7.2],[4.3, 6.2, 8.3]]) # defining x as input of train_loop which is 2 X 3 matrix !!
  tf.global_variables_initializer().run() # by running global variable initializer we initialize all the variable at once
  for i in xrange(len(preds)):
    print"{}:{}".format(i,preds[i].eval())
    
```

1 X 3 matrix each time is different !

**We hardcoded input value for training loop** , but we don't do it, 

**Place holder ** allows us to feed in value such as by reading from a text file!

```
import tensorflow as tf

a = tf.placeholder('float', None) # a is a placeholder, it will hold a scalar
b = a*4
print (a)
with tf.Session() as sess:
  print(sess.run(b, feed_dict = {a : [1,2,3]})) # feed_dict is how we are feeding the placeholder
```
